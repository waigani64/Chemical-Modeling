```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
K1 = 10**(-6.3)
K2 = 10**(-10.3)
KW = 10**(-14)

# Define pH range
pH = np.arange(0, 14.1, 0.1)

# Calculate carbonate species concentrations
H = 10**(-pH)
CO2 = 1 / (1 + (K1 / H) + (K1 * K2 / H**2))
HCO3 = K1 * CO2 / H
CO3 = K1 * K2 * CO2 / H**2
OH = KW / H
H2CO3 = H * CO2

# Plot results
plt.plot(pH, CO2, label='CO2')
plt.plot(pH, HCO3, label='HCO3-')
plt.plot(pH, CO3, label='CO3--')
plt.plot(pH, H2CO3, label='H2CO3')
plt.plot(pH, OH, label='OH-')
plt.xlabel('pH')
plt.ylabel('Concentration (M)')
plt.title('Carbonate Species Concentrations with pH')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define constants
K1 = 10**(-6.3)
K2 = 10**(-10.3)
KW = 10**(-14)

# Define pH range
pH = np.arange(0, 14.1, 0.1)

# Define temperature range
T = np.arange(0, 100, 1)

# Create a meshgrid of pH and temperature values
pH_mesh, T_mesh = np.meshgrid(pH, T)

# Calculate carbonate species concentrations for each pH and temperature value
H = 10**(-pH_mesh)
CO2 = 1 / (1 + (K1 / H) + (K1 * K2 / H**2))
HCO3 = K1 * CO2 / H
CO3 = K1 * K2 * CO2 / H**2
OH = KW / H
H2CO3 = H * CO2

# Plot results
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(pH_mesh, T_mesh, HCO3, cmap='coolwarm')
ax.set_xlabel('pH')
ax.set_ylabel('Temperature (C)')
ax.set_zlabel('C=HCO3 Concentration (M)')
plt.title('Carbonate Species Concentrations with pH and Temperature')
plt.show()
```
